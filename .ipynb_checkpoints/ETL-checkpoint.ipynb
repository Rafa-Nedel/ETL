```python
import pandas as pd
import pandera as pa
```

```python
tabela = pd.read_csv('ocorrencia.csv', sep=';', parse_dates=['ocorrencia_dia'], dayfirst=True)
tabela.head(10)
```

```python
schema = pa.DataFrameSchema (
    columns = {
        "codigo_ocorrencia":pa.Column(pa.Int)
    }
)
```

```python
schema.validate(tabela)
```
